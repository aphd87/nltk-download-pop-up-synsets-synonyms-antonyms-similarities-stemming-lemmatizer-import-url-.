```python
#Synsets

syn = wordnet.synsets("program")
print(syn[0].definition())
print(syn[0].name())
print(syn[0].lemmas()[0].name())
print(syn[0].examples())
```

```python
synonyms = []
for syn in wordnet.synsets('good'):  #There is a for-loop that is being used here.
    for lemma in syn.lemmas():
        synonyms.append(lemma.name())
print(synonyms)
```

```python
antonyms = []
for syn in wordnet.synsets("good"):
    for ant in syn.lemmas():
        if ant.antonyms():
            antonyms.append(ant.antonyms()[0].name())
print(antonyms)
```

```python
#Similarities

w1 = wordnet.synset('ship.n.01')
w2 = wordnet.synset('boat.n.01')
print(w1.wup_similarity(w2))
#Ship and boat are similar, therefore there is a high score here.
```

```python
#Import URL Lib Request
response = urllib.request.urlopen('https://www.espn.com')
html = response.read()
print(html)
```

```python
#conda install bs4
# conda install html5lib

soup = BeautifulSoup(html, "html.parser")
text = soup.get_text(strip=True)
print (text)
```

```python
tokens = [t for t in text.split()]
print(tokens)
#The text has been tokenized here
```

```python
#Frequency

freq = nltk.FreqDist(tokens)
for key, val in freq.items(): print (str(key) + ':' + str(val))
```

```python
#Stopwords

clean_tokens = tokens[:]
for token in tokens:
    if token in stopwords.words('english'):
        clean_tokens.remove(token)

freq = nltk.FreqDist(clean_tokens)
for key, val in freq.items(): print (str(key) + ':' + str(val))
    
#Now, stop words are gone.
```

```python
freq.plot(20, cumulative=False)
```

```python
#We can see which words are used the most here.

print (len(clean_tokens))
```
